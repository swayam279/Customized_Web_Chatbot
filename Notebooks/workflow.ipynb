1) User enters URL
2) The URL is validated by url_validator.py. This happens inside sitemap.py
3) The validated URL is sent to sitemap.py to create a sitemap and fetch all the URLs
4) All the URLs are sent to crawler.py in form of a list[URLs] and it asynchronously fetches the content of all URLs in batch(parallel) It returns a list[dict] with url, contents for each URL.
5) The resulting markdown is passed through markdown_cleaner.py to pre-process the markdown contents (This is integrated in the crawler.py file).
6) The list[dict] is then passed to the document_generator.py to generate a list[Document] with page_contents and metadata (url, base_url)
7) The documents containing page contents of each url are sent to text_splitter.py to split into chunks
8) The chunks are then passed to vector_store.py to store into chroma vector database if they were not already inside.
9) The retriever fetches data from this vector_store and passes data to the agent/LLM to answer user queries.
